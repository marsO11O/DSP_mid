```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import thinkdsp as tp
from scipy.signal import chirp, spectrogram
```

```python
dataframe=pd.read_csv("https://query1.finance.yahoo.com/v7/finance/download/%5EGSPC?period1=-1325635200&period2=1588550400&interval=1d&events=history")
```

```python
print(dataframe[:5],"\n")

print(dataframe[-5:])
```

```python
ys=dataframe['Close'].values
plt.plot(np.arange(len(ys)),ys)
```

```python
plt.plot(np.arange(len(ys[-5000:])),ys[-5000:])
```

```python
yWave= thinkdsp.Wave(ys, framerate= 10000)
yWave.plot()
```

```python
yWave.make_spectrum().make_integrated_spectrum().plot_power()
plt.xscale('log')
plt.yscale('log')
```
